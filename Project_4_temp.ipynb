```python
# 1. Define your classes (the order matters for the class_id)
classes = ["Cat", "Dog"]

# 2. Export the dataset directly to YOLO format
dataset.export(
    export_dir="/root/my_yolo_dataset",
    dataset_type=fo.types.YOLOv5Dataset, # Works for YOLOv5, v8, v10, v11
    label_field="ground_truth",
    classes=classes,
)
```

```python
# 1. Provide your paths
img_path = "/root/my_yolo_dataset/images/val/0007cebe1b2ba653.jpg"
label_path = "/root/my_yolo_dataset/labels/val/0007cebe1b2ba653.txt"

# 2. Load image
image = cv2.imread(img_path)
h, w, _ = image.shape

# 3. Read the YOLO .txt file
with open(label_path, "r") as f:
    lines = f.readlines()

for line in lines:
    parts = line.split()
    cls, cx, cy, bw, bh = map(float, parts)

    # Convert YOLO (center) to OpenCV (top-left/bottom-right pixels)
    # This is the "Inversion" of the math we discussed earlier
    x1 = int((cx - bw/2) * w)
    y1 = int((cy - bh/2) * h)
    x2 = int((cx + bw/2) * w)
    y2 = int((cy + bh/2) * h)

    # Draw the rectangle
    color = (0, 255, 0) if cls == 0 else (255, 0, 0) # Green for Cat, Blue for Dog
    cv2.rectangle(image, (x1, y1), (x2, y2), color, 3)
    clas = "Cat" if cls == 0 else "Dog"
    cv2.putText(image, f"Class {clas}", (x1, y1-10), cv2.FONT_HERSHEY_SIMPLEX, 1, color, 2)

plt.figure(figsize=(10,10))
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

```python
import os
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np

base_path = "/root/.cache/kagglehub/datasets/balraj98/duts-saliency-detection-dataset/versions/1"
train_img_dir = os.path.join(base_path, "DUTS-TR", "DUTS-TR-Image")
train_mask_dir = os.path.join(base_path, "DUTS-TR", "DUTS-TR-Mask")

# Get a sample filename
sample_filename = os.listdir(train_img_dir)[3]
# Note: Masks are often .png while images are .jpg;
# we swap the extension to find the matching mask
mask_filename = sample_filename.replace(".jpg", ".png")

# Load images
img = np.array(Image.open(os.path.join(train_img_dir, sample_filename)))
mask = np.array(Image.open(os.path.join(train_mask_dir, mask_filename)))
#img = Image.open(os.path.join(train_img_dir, sample_filename))
#mask = Image.open(os.path.join(train_mask_dir, mask_filename))
highlighted_img = img.copy()

# Anywhere the mask is bright (white), increase the Red channel
# (Assuming mask values are 0-255)
highlighted_img[mask > 128, 0] = 255  # Red channel
# Optional: lower the Green channel to make the pink "deeper"
highlighted_img[mask > 128, 1] = 150
highlighted_img[mask > 128, 2] = 255  # Blue channel

# Plotting
fig, ax = plt.subplots(1, 3, figsize=(16, 5))
ax[0].imshow(img)
ax[0].set_title("Original Image")
ax[1].imshow(mask)
ax[1].set_title("Saliency Mask (Ground Truth)")
ax[2].imshow(highlighted_img)
ax[2].set_title("Highlighted mask")
for a in ax: a.axis('off')
plt.show()
```

```python
import torch
from torch.utils.data import Dataset
from torchvision import transforms

class DUTSDataset(Dataset):
    def __init__(self, img_dir, mask_dir, transform=None):
        self.img_dir = img_dir
        self.mask_dir = mask_dir
        self.transform = transform
        # Ensure we only grab valid image files and sort for consistency
        self.images = sorted([f for f in os.listdir(img_dir) if f.endswith('.jpg')])

    def __len__(self):
        return len(self.images)

    def __getitem__(self, idx):
        img_name = self.images[idx]
        mask_name = img_name.replace(".jpg", ".png")

        img_path = os.path.join(self.img_dir, img_name)
        mask_path = os.path.join(self.mask_dir, mask_name)

        # Load as RGB and Grayscale respectively
        image = Image.open(img_path).convert("RGB")
        mask = Image.open(mask_path).convert("L")

        if self.transform:
            image = self.transform(image)
            mask = self.transform(mask)

        return image, mask

# Standard transforms: Resize and convert to Tensors
transform = transforms.Compose([
    transforms.Resize((256, 256)),
    transforms.ToTensor(),
])

# Initialize
train_dataset = DUTSDataset(train_img_dir, train_mask_dir, transform=transform)
train_loader = torch.utils.data.DataLoader(train_dataset, batch_size=16, shuffle=True)
```

```python
import segmentation_models_pytorch as smp

model = smp.Unet(
    encoder_name="resnet34",        # A solid, reliable backbone
    encoder_weights="imagenet",     # Use pre-trained weights to speed up learning
    in_channels=3,                  # RGB input
    classes=1,                      # 1 output channel (the saliency mask)
    activation='sigmoid'            # Squashes output between 0 and 1
)
```

```python
import torch.optim as optim
import torch.nn as nn

# Move model to GPU if available
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)

# Loss Function: Binary Cross Entropy with Logits (stable Sigmoid + BCE)
#small_indices = range(100)
#train_subset = torch.utils.data.Subset(train_dataset, small_indices)

# Use this loader instead to see if an epoch finishes in seconds
#train_loader = torch.utils.data.DataLoader(train_subset, batch_size=8, shuffle=True)

criterion = nn.BCELoss()

# Optimizer
optimizer = optim.Adam(model.parameters(), lr=1e-4)

epochs = 10
for epoch in range(epochs):
    model.train()
    running_loss = 0.0

    for images, masks in train_loader:
        images = images.to(device)
        masks = masks.to(device)

        # 1. Clear old gradients from the previous batch
        optimizer.zero_grad()

        # 2. Forward pass: Make a prediction
        outputs = model(images)

        # 3. Calculate Loss (how far off are we?)
        loss = criterion(outputs, masks)

        # 4. Backward pass: Calculate the gradients
        loss.backward()

        # 5. Optimization: Update the weights based on gradients
        optimizer.step()

        running_loss += loss.item()

    epoch_loss = running_loss / len(train_loader)
    print(f"Epoch [{epoch+1}/{epochs}], Loss: {epoch_loss:.4f}")

print("Training Complete!")
```
