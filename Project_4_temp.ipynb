```python
import fiftyone as fo
import fiftyone.zoo as foz

dataset = foz.load_zoo_dataset(
              "open-images-v6",
              split="validation",
              label_types=["detections"],
              classes=["Cat", "Dog"],
              max_samples=100,
          )
```

```python
import matplotlib.pyplot as plt
import cv2

sample = dataset.first()
#sample = list(dataset)[4]
image = sample.filepath

# Load and display
image = cv2.imread(image)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.imshow(image)
plt.title("Title")
plt.show()
```

```python
# Print the bounding box data
for detection in sample.ground_truth.detections:
    print(f"Label: {detection.label}")
    print(f"Bounding Box [x, y, w, h]: {detection.bounding_box}")
```

```python
# 1. Define your classes (the order matters for the class_id)
classes = ["Cat", "Dog"]

# 2. Export the dataset directly to YOLO format
dataset.export(
    export_dir="/root/my_yolo_dataset",
    dataset_type=fo.types.YOLOv5Dataset, # Works for YOLOv5, v8, v10, v11
    label_field="ground_truth",
    classes=classes,
)
```

```python
# 1. Provide your paths
img_path = "/root/my_yolo_dataset/images/val/0007cebe1b2ba653.jpg"
label_path = "/root/my_yolo_dataset/labels/val/0007cebe1b2ba653.txt"

# 2. Load image
image = cv2.imread(img_path)
h, w, _ = image.shape

# 3. Read the YOLO .txt file
with open(label_path, "r") as f:
    lines = f.readlines()

for line in lines:
    parts = line.split()
    cls, cx, cy, bw, bh = map(float, parts)

    # Convert YOLO (center) to OpenCV (top-left/bottom-right pixels)
    # This is the "Inversion" of the math we discussed earlier
    x1 = int((cx - bw/2) * w)
    y1 = int((cy - bh/2) * h)
    x2 = int((cx + bw/2) * w)
    y2 = int((cy + bh/2) * h)

    # Draw the rectangle
    color = (0, 255, 0) if cls == 0 else (255, 0, 0) # Green for Cat, Blue for Dog
    cv2.rectangle(image, (x1, y1), (x2, y2), color, 3)
    clas = "Cat" if cls == 0 else "Dog"
    cv2.putText(image, f"Class {clas}", (x1, y1-10), cv2.FONT_HERSHEY_SIMPLEX, 1, color, 2)

plt.figure(figsize=(10,10))
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```
